**Taking necessary columns only**

```python
new_df = df[['id', 'name', 'host_id', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type', 'price', 'minimum_nights', 'number_of_reviews', 'calculated_host_listings_count', 'availability_365']]
new_df.head(5)
```

# 1. **What can we learn about different hosts and areas?**







```python
areas_hosts = new_df.groupby(['host_name', 'neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
areas_hosts.sort_values(by = 'calculated_host_listings_count', ascending = False).head(5)
```

The most number are from Manhatan created Sonder(NYC), Blueground.

# 2. What can we learn from predictions?(ex: locations, prices, reviews, etc)


```python
reviews_areas = new_df.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
reviews_areas
```

```python
area = reviews_areas['neighbourhood_group']
review = reviews_areas['number_of_reviews']
fig = plt.figure(figsize = (12,7))
plt.bar(area, review, color = 'blue', width = 0.5)
plt.xlabel('area')
plt.ylabel('review')
plt.title('Area vs number of shows')
plt.show()

```

```python
area_price = new_df.groupby(['price'])['number_of_reviews'].max().reset_index()
area_price.head(5)
```

```python
area = area_price['price']
price = area_price['number_of_reviews']
fig = plt.figure(figsize = (14,7))
plt.scatter(area, price, color ='indigo')
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.title('Price vs Number Of Reviews')
plt.show()
```

From the above data we can see that most people prefer to stay in less price place.

# 3. Which hosts are the bussiest and why?

```python
bussiest_hosts = new_df.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
bussiest_hosts = bussiest_hosts.sort_values(by = 'number_of_reviews', ascending = False).head(10)
bussiest_hosts
```

```python
name = bussiest_hosts['host_name']
reviews = bussiest_hosts['number_of_reviews']
fig = plt.figure(figsize = (12,6))
plt.bar(name, reviews, color = 'yellow', width = 0.6)
plt.xlabel("Name Of The Hosts")
plt.ylabel("Number Of Reviews")
plt.title("Bussiest Hosts")
plt.show()
```

# **4. Is there any noticeable difference of traffic among different areas and what coulf be the reason for it?**

```python
traffic_areas = new_df.groupby(['neighbourhood_group', 'room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by = 'minimum_nights', ascending = False)
traffic_areas
```

```python
room_type = traffic_areas['room_type']
stayed = traffic_areas['minimum_nights']
fig = plt.figure(figsize = (12,6))
plt.bar(room_type, stayed, color = 'purple', width = 0.3)
plt.xlabel("Room Type")
plt.ylabel("Minimum Nights")
plt.title("Traffic Areas")
plt.show()
```
